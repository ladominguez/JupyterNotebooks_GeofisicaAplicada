# Misión
En este cuaderno de Jupyter usaremos la libreria **Pyrocko** para estimar los tiempo de propagación utilizando la Ecuación Eikonal. Al final de la práctica podrás modelar el resultado de introducir anomalías sísmicas en el modelo de velocidades.
## 1. Inicialización del modelo de velocidades
En la siguiente celda se carga el modelo de velocidades y se inicializa las matrices que contienen el modelos de velocidades. También se determina la localización de la fuente. En este caso se va a crear una matriz de **$1500\times500$** que contiene una malla en la que vamos a asignar un valor a cada capa. La fuente se va a localizar en la posición en supercifie **$x=0.0$** y a una profundidad de **$25km$**

```python
import numpy as num
from matplotlib import pyplot as plt
from pyrocko.modelling import eikonal

km = 1000.
nx, ny = 1500, 500                # Tamaño de la malla
delta = 90*km / float(nx)         # Espaciamiento
source_x, source_y = 0.0, 25*km   # Localización de la fuente

# Indexing arrays
x = num.arange(nx) * delta - 2*km
y = num.arange(ny) * delta
x2 = x[num.newaxis, :]
y2 = y[:, num.newaxis]
```

## 2. Define el modelo de velocidades
En la siguiente celda vamos a crear un modelo de **5 capas** con velocidades de $2.5$, $3.5$, $5$, y $6km/s$.

```python
speeds = num.ones((ny, nx))
nlayer = ny // 5
speeds[0*nlayer:1*nlayer, :] = 2500.
speeds[1*nlayer:2*nlayer, :] = 3500.
speeds[2*nlayer:3*nlayer, :] = 5000.
speeds[3*nlayer:4*nlayer, :] = 6000.
speeds[4*nlayer:, :] = 8000.
```

# 3. Inicialización de los tiempos
Para establecer las condiciones iniciales de la matriz de tiempos de arribo se debe de crear una matriz con las mismas dimensiones que la matriz de velocidades y posteriormente asignar el valor de **cero**, en el **nodo** de la fuente. Todos los demás deben de ser menores negativos ($t=-1$).

```python
times = num.zeros((ny, nx)) - 1.0
times[int(round(source_y/delta)), int(round((source_x-x[0])//delta))] = 0.0
```

# 4. Ejecuta el programa que resuelve la ecuación Eikonal. 
Con un solo comando puedes resolver la ecuación Eikonal. Esta función recibe como parámetros de entrada el modelo de velocidades, la matriz de tiempos de viaje y el valor de espaciamiento de puntos entre la malla. 

```python
eikonal.eikonal_solver_fmm_cartesian(speeds, times, delta)
```

# 5. Grafica tus resultados 
Por último grafica tus resultados.

```python
%matplotlib inline
# Plot
fig = plt.figure(figsize=(12.0, 6.0))

axes = fig.add_subplot(1, 1, 1, aspect=1.0)
#axes.contourf(x/km, y/km, times, cmap='hot',alpha=0.2)
axes.invert_yaxis()
axes.contourf(x/km, y/km, speeds, alpha=0.4, cmap='gray')
axes.plot(source_x/km, source_y/km, '*', ms=20, color='white')
axes.set_xlabel('Distancia [km]')
axes.set_ylabel('Profundidad [km]')
axes.contour(x/km, y/km, times, colors='black', linestyles='dashed', linewidths=0.5)
cbar_ax = fig.add_axes([0.125, 0.1, 0.775, 0.025])
cbar = fig.colorbar(axes.contourf(x/km, y/km, times, cmap='hot', alpha=0.5), cax=cbar_ax, orientation='horizontal')
cbar.set_label('Tiempo de viaje [s]')
# fig.savefig('eikonal_example1.png')
plt.show()
```
